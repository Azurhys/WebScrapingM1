```python
def nbOffers(driver):
    try:
       
        job_offers_element = driver.find_element_by_class_name("sc-fzXfLV")
        
       
        job_offers_text = job_offers_element.text
        
        
        nb_offers = int(job_offers_text.split()[0])  
        
        return nb_offers
        
    except Exception as e:
        print("An error occurred in NB_OFFER:", str(e))
        return 0  
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def MainPage(driver, url):
    driver.get(url)
    # Wait for the page to load completely
    time.sleep(5)  # Adjust the sleep time as needed

def nbOffers(driver):
    try:
        # Locate the elements containing the job offers
        job_offers_elements = driver.find_elements(By.CLASS_NAME, "sc-bXCLTC")

        # Get the number of job offers
        nb_offers = len(job_offers_elements)
        
        return nb_offers
        
    except Exception as e:
        print("An error occurred in NB_OFFER:", str(e))
        return 0  # Or handle the exception as needed

# Example usage:
url = "https://www.welcometothejungle.com/fr/jobs?page=1&configure%5Bfilters%5D=website.reference%3Awttj_fr&configure%5BhitsPerPage%5D=30&aroundQuery=France&refinementList%5Boffice.country_code%5D%5B%5D=FR&refinementList%5Bcontract_type_names.fr%5D%5B%5D=CDI&refinementList%5Bcontract_type_names.fr%5D%5B%5D=Stage&query=%22data%20analyst%22&range%5Bexperience_level_minimum%5D%5Bmin%5D=0&range%5Bexperience_level_minimum%5D%5Bmax%5D=1"

driver = webdriver.Chrome()  # Create a Chrome WebDriver instance
MainPage(driver, url)
nb_offers = nbOffers(driver)
print("Number of job offers:", nb_offers)

# Close the browser window
driver.quit()
```
